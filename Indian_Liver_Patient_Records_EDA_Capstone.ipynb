**<span style="color:Teal;">Total proteins in the blood consist mainly of albumin and globulin. In the liver, protein production is a key function, as it synthesizes these proteins, which are crucial for maintaining blood volume, immune function, and the transport of substances. Abnormal levels of total proteins can indicate liver dysfunction or disease, as the liver may not be producing proteins adequately.
A decrease in total proteins may suggest liver disease such as cirrhosis or hepatitis, while increased levels may indicate dehydration or chronic inflammation. Monitoring these levels can provide valuable insight into liver health and its ability to synthesize proteins properly.**

We will be breaking down total proteins into two segments, Albumin and Albumin-Globulin ration

**<span style="color:Teal;">Albumin: Albumin(ALB) is a protein produced by the liver and plays a crucial role in maintaining the oncotic pressure of the blood, helping to keep fluid from leaking out of blood vessels. It also serves as a carrier for hormones, vitamins, and drugs. The normal range is between 3.5-5.0g/dL**

```python
# Clean up outliers
lower_percentile = df['Albumin'].quantile(0.01)  
upper_percentile = df['Albumin'].quantile(0.99)

# Replace outliers
df['Albumin'] = df['Albumin'].clip(lower=lower_percentile, upper=upper_percentile)

# Mean
mean_Albumin = np.mean(df['Albumin'])

# Median
median_Albumin = np.median(df['Albumin'])

# IQR
Albumin_q1 = np.percentile(df['Albumin'], 25)  # 25th percentile (Q1)
Albumin_q3 = np.percentile(df['Albumin'], 75)  # 75th percentile (Q3)
Albumin_iqr = Albumin_q3 - Albumin_q1

# Standard Deviation
Albumin_std_dev = np.std(df['Albumin'], ddof=1)

# Variance 
Albumin_variance = np.var(df['Albumin'], ddof=1)

# Range
Albumin_data_range = max(df['Albumin']) - min(df['Albumin'])

# IQR range
Albumin_iqr_range = (Albumin_q1, Albumin_q3)

print("The Albumin mean is:", mean_Albumin)
print("The Albumin median is:", median_Albumin)
print("The Albumin IQR is:", Albumin_iqr)
print("The Albumin STD is:", Albumin_std_dev)
print("The Albumin variance is:", Albumin_variance)
print("The Albumin range is:", Albumin_data_range)
print("The Albumin IQR range is:", Albumin_iqr_range)
```

#**<span style="color:Teal;">albumin-globulin (A/G) ratio: The albumin-globulin (A/G) ratio, which compares the levels of albumin and globulin in the blood, is an important marker of liver function since both proteins are produced by the liver. A low A/G ratio is often linked to liver diseases like cirrhosis or hepatitis, where liver damage leads to reduced albumin production and increased globulins due to inflammation. A high A/G ratio, though less common, can indicate conditions such as dehydration or protein malnutrition, where albumin levels are higher than globulins. A normal A/G ratio typically ranges from 1.1 to 2.5**
